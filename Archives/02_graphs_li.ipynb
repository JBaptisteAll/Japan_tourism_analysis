```python
mapping_interest = {
    "Not important at all": 0,
    "Slightly important": 1,
    "Moderately important": 2,
    "Very important": 3,
    "Essential": 4,
}
```

```python
columns = [
    'rating_interest_culture_and_history', 
    'rating_interest_food', 
    'rating_interest_nature_hiking', 
    'rating_interest_shopping_and_techno',
    'rating_interest_events_and_festivals', 
    'rating_interest_wellness',
    'rating_interest_theme_park'
]
```

```python
df_numeric = df.copy()

for col in columns:
    df_numeric[col] = df_numeric[col].map(mapping_interest)
```

```python
label_map = {
    "rating_interest_culture_and_history": "Culture & History",
    "rating_interest_food": "Food",
    "rating_interest_nature_hiking": "Nature & Hiking",
    "rating_interest_shopping_and_techno": "Shopping & Tech",
    "rating_interest_events_and_festivals": "Events & Festivals",
    "rating_interest_wellness": "Wellness",
    "rating_interest_theme_park": "Theme Parks",
}
```

```python

summary_df = (
    df_numeric[columns]
    .mean()
    .reset_index()
    .rename(columns = {
        "index": "interest",
        0: "value"
    })
)

summary_df["text_display"] = summary_df["value"].round(2)
summary_df["interest"] = summary_df["interest"].map(label_map)


fig_1 = px.bar (
    summary_df,
    x = "interest",
    y = "value",
    text = "text_display",
    title = "Un int√©r√™t fort‚Ä¶ mais une conversion faible",
)

fig_1.update_traces (
    textposition = "outside"
)

fig_1.update_layout (
    width=800,
    height=500,
    xaxis_title = "",
    yaxis_title = "Score moyen d‚Äôint√©r√™t (0‚Äì4)"
)
```

```python
essential_pct = []

for col in columns:
    pct = (df_numeric[col] == 4).mean() * 100
    essential_pct.append({
        "interest": col.replace("rating_interest_", "").replace("_", " ").title(),
        "essential_pct": round(pct, 1)
    })

essential_df = pd.DataFrame(essential_pct).sort_values(
    "essential_pct", ascending=False
)

essential_df
```

```python
essential_df["text_display"] = essential_df["essential_pct"].astype(str) + "%"


essential_df = essential_df.sort_values(
    "essential_pct", ascending=False
).reset_index(drop=True)

essential_df.loc[0, "text_display"] = "<b>" + essential_df.loc[0, "text_display"] + "</b>"
```

```python
#essential_df["interest"] = essential_df["interest"].map(label_map)
mapping_interest = {
    "Food": "Gastronomie",
    "Nature Hiking": "Nature & randonn√©e",
    "Culture And History": "Culture & patrimoine",
    "Wellness": "Bien-√™tre",
    "Shopping And Techno": "Shopping & technologie",
    "Events And Festivals": "√âv√©nements & festivals",
    "Theme Park": "Parcs √† th√®me"
}

essential_df["interest"] = (essential_df["interest"].map(mapping_interest))
```

```python
fig = px.bar(
    essential_df,
    x="interest",
    y="essential_pct",
    text="text_display",
    title="Crit√®res jug√©s ESSENTIELS par les voyageurs int√©ress√©s par le Japon (%)",
    color = "interest",
    color_discrete_map = {
        "Gastronomie": "#E6FF01",
        "Nature & randonn√©e": "#E9A107",
        "Culture & patrimoine": "#E9A107",
        "Bien-√™tre": "#0C0241",
        "Shopping & technologie": "#0C0241",
        "√âv√©nements & festivals": "#313131",
        "Parcs √† th√®me": "#313131"
        }
)

fig.update_traces(textfont = dict(size=15), textposition="outside")
fig.update_layout(
    width=800,
    height=500,
    xaxis_title="",
    yaxis_title="Part des r√©pondants (%)",
    yaxis_range=[0, max(essential_df["essential_pct"]) * 1.2],
    showlegend = False
)

fig.show()
```
